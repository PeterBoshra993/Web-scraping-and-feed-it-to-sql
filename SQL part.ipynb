```python
import sqlite3
conn = sqlite3.connect('scrapedbooks.db')
curo = conn.cursor()
```

```python
def get_rows(sql, data):
    global conn
    global curo
    curo.execute(sql, data)
    rows = curo.fetchall()
    for NAME, PRICE, RATING, Category_Id in rows:
        print(f"[title:{NAME} ----- price:{PRICE} £----- rates:{RATING} ----- Category_id: {Category_Id}]")
```

#### Write a sql query to get books that has rate >=3 and has 'Mr' in its name.

```python
get_rows("SELECT * FROM books WHERE RATING >= ? AND NAME LIKE ?", [3, '%Mr%'])
```

#### Write a sql query to get books that has rate >=3 or price > 20.

```python
get_rows("SELECT * FROM books WHERE RATING >= ? AND PRICE > ?",[3, 20])
```
